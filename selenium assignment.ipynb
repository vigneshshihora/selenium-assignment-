```python

def scrape_shine_jobs(job_title, location, num_jobs=10):
    
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

    # Navigate to the Shine website
    driver.get("https://www.shine.com/")

    try:
        
        job_title_field = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "qsb-keyword-sugg"))
        )
        job_title_field.send_keys(job_title)

        location_field = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "qsb-location-sugg"))
        )
        location_field.clear()
        location_field.send_keys(location)

        # Click the search button
        search_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "search_btn"))
        )
        search_button.click()

        
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "sjsresult"))
        )

        # Scrape job data
        job_titles = []
        job_locations = []
        company_names = []
        experience_required = []

        job_listings = driver.find_elements(By.CLASS_NAME, "sjsresult")[:num_jobs]
        for listing in job_listings:
            job_titles.append(listing.find_element(By.CLASS_NAME, "sjs__job-title").text.strip())
            job_locations.append(listing.find_element(By.CLASS_NAME, "sjs__job-location").text.strip())
            company_names.append(listing.find_element(By.CLASS_NAME, "sjs__company-name").text.strip())
            experience_required.append(listing.find_element(By.CLASS_NAME, "sjs__job-experience").text.strip())

        # Create DataFrame
        data = {
            'Job Title': job_titles,
            'Job Location': job_locations,
            'Company Name': company_names,
            'Experience Required': experience_required
        }
        df = pd.DataFrame(data)

    finally:
        
        driver.quit()

    return df


job_title = "Data Scientist"
location = "Bangalore"
num_jobs = 10
df = scrape_shine_jobs(job_title, location, num_jobs)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_flipkart_reviews(product_url, num_reviews=100):
    # Configure Selenium
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  # Optional: Run in headless mode (no browser window)
    driver = webdriver.Chrome(options=options)

    # Navigate to the product reviews page
    driver.get(product_url)

    try:
        # Click on "See all reviews" to view all reviews
        see_all_reviews = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//div[@class='_3UAT2v _16PBlm']"))
        )
        see_all_reviews.click()

        # Wait for reviews to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "_2c2kV-"))
        )

        # Scrape review data
        ratings = []
        review_summaries = []
        full_reviews = []

        reviews = driver.find_elements(By.CLASS_NAME, "_1AtVbE")
        for review in reviews[:num_reviews]:
            # Get rating
            rating = review.find_element(By.XPATH, ".//div[@class='_3LWZlK _1BLPMq']").text.strip()
            ratings.append(rating)

            # Get review summary
            review_summary = review.find_element(By.XPATH, ".//p[@class='_2-N8zT']").text.strip()
            review_summaries.append(review_summary)

            # Get full review
            full_review = review.find_element(By.XPATH, ".//div[@class='t-ZTKy']").text.strip()
            full_reviews.append(full_review)

    finally:
        # Close the browser
        driver.quit()

    # Create DataFrame
    data = {
        'Rating': ratings,
        'Review Summary': review_summaries,
        'Full Review': full_reviews
    }
    df = pd.DataFrame(data)

    return df

# URL of the product reviews page
product_url = "https://www.flipkart.com/apple-iphone-15-black-128-gb/p/itm6ac6485515ae4?pid=MOBGTAGPTB3VS24W&lid=LSTMOBGTAGPTB3VS24WVZNSC6&marketplace=FLIPKART&q=iphone+15&store=tyy/4io&srno=s_1_2&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&fm=organic&iid=b596c120-8114-44e8-9819-0b4dee130fa9.MOBGTAGPTB3VS24W.SEARCH&ppt=hp&ppn=homepage&ssid=s3c8i7dsvk0000001710403029549&qH=2f54b45b321e3ae5"

# Scrape reviews
df = scrape_flipkart_reviews(product_url, num_reviews=100)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_flipkart_sneakers(num_sneakers=100):
   
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

    # Navigate to Flipkart and search for sneakers
    driver.get("https://www.flipkart.com/")
    search_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME, "q"))
    )
    search_field.send_keys("sneakers")
    search_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "L0Z3Pu"))
    )
    search_button.click()

    try:
       
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div._2B099V"))
        )

        # Scrape sneaker data
        brands = []
        descriptions = []
        prices = []

        sneakers = driver.find_elements(By.CLASS_NAME, "div._2B099V")[:num_sneakers]
        for sneaker in sneakers:
            # Get brand
            brand = sneaker.find_element(By.CLASS_NAME, "div._2WkVRV").text.strip()
            brands.append(brand)

            # Get description
            description = sneaker.find_element(By.CLASS_NAME, "aIRyel").text.strip()
            descriptions.append(description)

            # Get price
            price = sneaker.find_element(By.CLASS_NAME, "._30jeq3").text.strip()
            prices.append(price)

    finally:
      
        driver.quit()

    # Create DataFrame
    data = {
        'Brand': brands,
        'Description': descriptions,
        'Price': prices
    }
    df = pd.DataFrame(data)

    return df

# Scrape sneakers
df = scrape_flipkart_sneakers(num_sneakers=100)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_amazon_laptops(num_laptops=10):
    
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

    # Navigate to Amazon and search for laptops
    driver.get("https://www.amazon.in/")
    search_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "twotabsearchtextbox"))
    )
    search_field.send_keys("Laptop")
    search_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "nav-search-submit-button"))
    )
    search_button.click()

    # Filter by CPU Type to "Intel Core i7"
    cpu_type_filter = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//span[text()='Intel Core i7']"))
    )
    cpu_type_filter.click()

    try:
        
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, "//div[@data-component-type='s-search-result']"))
        )

        # Scrape laptop data
        titles = []
        ratings = []
        prices = []

        laptops = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")[:num_laptops]
        for laptop in laptops:
            # Get title
            title = laptop.find_element(By.XPATH, ".//h2/a").text.strip()
            titles.append(title)

            # Get rating
            try:
                rating = laptop.find_element(By.XPATH, ".//span[@class='a-icon-alt']").get_attribute("textContent").split()[0]
            except:
                rating = "Not Available"
            ratings.append(rating)

            # Get price
            try:
                price = laptop.find_element(By.XPATH, ".//span[@class='a-price-whole']").text.strip()
            except:
                price = "Not Available"
            prices.append(price)

    finally:
       
        driver.quit()

    # Create DataFrame
    data = {
        'Title': titles,
        'Ratings': ratings,
        'Price': prices
    }
    df = pd.DataFrame(data)

    return df

# Scrape laptops
df = scrape_amazon_laptops(num_laptops=10)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_top_quotes(num_quotes=1000):
    
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

    # Navigate to AzQuotes.com
    driver.get("https://www.azquotes.com/")

    try:
       
        top_quote_link = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.LINK_TEXT, "Top Quotes"))
        )
        top_quote_link.click()

        #
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "wrap"))
        )

        # Scrape quote data
        quotes = []
        authors = []
        types = []

        quote_elements = driver.find_elements(By.CLASS_NAME, "wrap")[:num_quotes]
        for quote_element in quote_elements:
            # Get quote
            quote = quote_element.find_element(By.CLASS_NAME, "title").text.strip()
            quotes.append(quote)

            # Get author
            author = quote_element.find_element(By.CLASS_NAME, "author").text.strip()
            authors.append(author)

            # Get type of quote
            quote_type = quote_element.find_element(By.CLASS_NAME, "kw-box").text.strip()
            types.append(quote_type)

    finally:
       
        driver.quit()

    # Create DataFrame
    data = {
        'Quote': quotes,
        'Author': authors,
        'Type': types
    }
    df = pd.DataFrame(data)

    return df

# Scrape top quotes
df = scrape_top_quotes(num_quotes=1000)
print(df)
```

```python
from selenium import webdriver
import pandas as pd

def scrape_prime_ministers():
   
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

    #
    driver.get("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")

    try:
        # Scrape data using class names
        names = []
        born_dead = []
        term_of_office = []
        remarks = []

        # Find table containing Prime Ministers data
        table = driver.find_element_by_class_name('table')
        rows = table.find_elements_by_tag_name('tr')

        # Loop through rows of the table
        for row in rows[1:]: # Skipping the header row
            # Extract data from each cell in the row
            cells = row.find_elements_by_tag_name('td')
            name = cells[0].text.strip()
            born_dead_info = cells[1].text.strip()
            term_of_office_info = cells[2].text.strip()
            remark = cells[3].text.strip()

            # Splitting born_dead_info to separate born and dead dates
            born_dead_info_parts = born_dead_info.split('-')
            born = born_dead_info_parts[0].strip()
            dead = born_dead_info_parts[1].strip() if len(born_dead_info_parts) > 1 else None

            # Splitting term_of_office_info to separate start and end dates
            term_of_office_info_parts = term_of_office_info.split('-')
            start_date = term_of_office_info_parts[0].strip()
            end_date = term_of_office_info_parts[1].strip() if len(term_of_office_info_parts) > 1 else None

            # Append data to respective lists
            names.append(name)
            born_dead.append((born, dead))
            term_of_office.append((start_date, end_date))
            remarks.append(remark)

    finally:
       
        driver.quit()

   
    data = {
        'Name': names,
        'Born-Dead': born_dead,
        'Term of Office': term_of_office,
        'Remarks': remarks
    }
    df = pd.DataFrame(data)
    return df

# Scrape and display Prime Ministers data
prime_ministers_df = scrape_prime_ministers()
print(prime_ministers_df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_expensive_cars():
    
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)

   
    driver.get("https://www.motor1.com/")

    try:
        # Type in the search bar
        search_bar = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.NAME, "q"))
        )
        search_bar.send_keys("50 most expensive cars")
        search_bar.submit()

        # Click on the search result for "50 most expensive cars in the world"
        search_result = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.LINK_TEXT, "50 Most Expensive Cars In The World"))
        )
        search_result.click()

       
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "lazyload"))
        )

        # Scrape car data
        cars = []
        prices = []

        car_elements = driver.find_elements(By.CLASS_NAME, "lazyload")
        for car_element in car_elements[:50]:  # Limiting to 50 cars
            # Get car name
            car_name = car_element.find_element(By.TAG_NAME, "h2").text.strip()
            cars.append(car_name)

            # Get car price
            try:
                price = car_element.find_element(By.CLASS_NAME, "price").text.strip()
            except:
                price = "Price not available"
            prices.append(price)

    finally:
       
        driver.quit()

    # Create DataFrame
    data = {
        'Car Name': cars,
        'Price': prices
    }
    df = pd.DataFrame(data)
    return df

# Scrape and display the list of 50 most expensive cars
expensive_cars_df = scrape_expensive_cars()
print(expensive_cars_df)
```
